```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mei = pd.read_csv("../data/MEI_CLI_05072020232700668.csv")
mei.Subject.value_counts()

# BCI - Business Confidence Index
# CCI - Consumer Confidence Index

BCI = mei.loc[0:76][['Subject', 'TIME','Value']]

BCI.plot("TIME", "Value", figsize = (12, 3), title="BCI")
plt.ylim(94, 106)
plt.show()

mei_CCI = mei.loc[77:]

CCI = mei_CCI[['Subject', 'TIME','Value']].reset_index(drop=True)

# CCI.to_csv('../data/CCI_2014-2020.csv')
# BCI.to_csv('../data/BCI_2014-2020.csv')

CCI.plot("TIME", "Value", figsize = (12, 3), title="CCI")
plt.ylim(94, 106)
plt.show()

# Business and consumer confidence in Brazil decreased due to 2014 economic crisis and then 
# was slowly recovering until end 2019. The most recent dip concides with the pandemic.
```

```python
indic = BCI
indic['BCI'] = indic['Value']
indic['CCI'] = CCI['Value']
indic.drop(columns=['Subject', 'Value'], inplace=True)
indic
```

```python
indicators = pd.read_csv("../data/Book1.csv", sep=";")
indicators.drop(columns=["Country", "Time"], inplace=True)

indicators = indicators.apply(lambda x: pd.to_datetime(x, errors='coerce').dt.strftime('%Y-%m') if x.name in ['TIME'] else x)

indicators.Subject.value_counts(60)
indicators
```

```python
# Reference period 2015=100

col = 'Sales > Retail trade > Total retail trade > Volume'

retail_trade_index = indicators.loc[(indicators['Subject'] == col)].reset_index(drop=True)
# retail_trade_index.head()

indic['retail_trade_index'] = retail_trade_index['Value']
```

```python
# Reference period 2015=100

col = "Consumer Price Index > Clothing and footwear (COICOP 03) > Total > Total"

CPI_clothing_index = indicators.loc[(indicators['Subject'] == col)&(indicators['Unit'] == 'Index')].reset_index(drop=True)
CPI_clothing_index.head()
indic['CPI_clothing_index'] = CPI_clothing_index['Value']
```

```python
CPI_index.plot()
plt.show()
```

```python
# Reference period 2015=100

col = 'Labour Compensation > Earnings > All activities > Monthly'

labour_earnings_index = indicators.loc[(indicators['Subject'] == col)&(indicators['Unit'] == 'Index')].reset_index(drop=True)

# labour_earnings_index.head()

indic['labour_earnings_index'] = labour_earnings_index['Value']
```

```python
indic.plot(x="TIME", figsize = (12, 4))
plt.show()
```
